```python
import dash
import dash_leaflet as dl
import pandas as pd
from dash import html, dcc, Output, Input
from pyproj import Transformer
from shapely.geometry import MultiPolygon, Polygon
import plotly.express as px
import numpy as np

transformer = Transformer.from_crs("EPSG:3857", "EPSG:4326", always_xy=True)

county_options = MDTraffic['COUNTY_DES'].unique()

app = dash.Dash(__name__)

app.layout = html.Div([
    html.Div([
        html.Div([
            html.Label("County"),
            dcc.Dropdown(id='county-dropdown',
                         options=[{'label': c, 'value': c} for c in county_options],
                         value='MONTGOMERY', clearable=False),
        ], style={'width': '20%', 'display': 'inline-block', 'padding-right': '10px'}),

        html.Div([
            html.Label("Street Name"),
            dcc.Dropdown(id='street-dropdown', multi=True),
        ], style={'width': '40%', 'display': 'inline-block', 'padding-right': '10px'}),

        html.Div([
            html.Label("Color Polygons By"),
            dcc.Dropdown(id='color-dropdown', clearable=False, value='drainagecl',
                         options=[
                             {'label': 'Component Name', 'value': 'compname'},
                             {'label': 'Drainage Class', 'value': 'drainagecl'},
                             {'label': 'Hydrologic Group', 'value': 'hydgrp'},
                             {'label': 'Corrosion Concrete', 'value': 'corcon'}
                         ]),
        ], style={'width': '35%', 'display': 'inline-block'}),
    ], style={'padding': '10px'}),

    dl.Map([
        dl.TileLayer(),
        dl.LayerGroup(id="layers")
    ], style={'width': '100%', 'height': '600px'}, id="map", zoom=10, center=(39.1547, -77.2405))
])

@app.callback(
    Output('street-dropdown', 'options'),
    Output('street-dropdown', 'value'),
    Input('county-dropdown', 'value')
)
def update_street_dropdown(county):
    streets = MDTraffic[MDTraffic['COUNTY_DES'] == county]['ROADNAME'].dropna().unique()[:100]
    return [{'label': s, 'value': s} for s in streets], []

@app.callback(
    Output("layers", "children"),
    Input('county-dropdown', 'value'),
    Input('street-dropdown', 'value'),
    Input('color-dropdown', 'value')
)
def update_map(county, selected_streets, color_by):
    traffic_df = MDTraffic[MDTraffic['COUNTY_DES'] == county]
    if selected_streets:
        traffic_df = traffic_df[traffic_df['ROADNAME'].isin(selected_streets)]
    traffic_4326 = traffic_df.to_crs(MDSoilPoly.crs)
    traffic_union = traffic_4326.unary_union

    polylines = [
        dl.Polyline(
            positions=[transformer.transform(x, y)[::-1] for x, y, *_ in row.geometry.coords],
            color='blue',
            children=dl.Popup(html.Div([
                html.B("Road Name: "), row['ROADNAME'], html.Br(),
                html.B("AADT 2022: "), row['AADT_2022']
            ]))
        ) for _, row in traffic_df.iterrows()
    ]

    soils = MDSoilPoly[MDSoilPoly.geometry.intersects(traffic_union)].copy()
    comps = MDSoilComp[MDSoilComp['majcompflag'] == 'Yes'][['mukey', 'compname', 'compkind', 'drainagecl', 'hydgrp', 'corcon', 'taxclname']]
    soils['mukey'] = soils['MUKEY'].astype(str)
    comps['mukey'] = comps['mukey'].astype(str)
    soils = soils.merge(comps, on='mukey', how='left')

    # Color logic
    def get_color(row):
        if color_by == 'drainagecl':
            drainage_order = ['Excessively drained', 'Somewhat excessively drained', 'Well drained', 
                              'Moderately well drained', 'Somewhat poorly drained', 'Poorly drained',
                              'Very poorly drained', 'Subaqueous']
            colors = px.colors.sequential.Plasma_r
            drainage_colors = {k: colors[int(i*(len(colors)-1)/(len(drainage_order)-1))] 
                               for i, k in enumerate(drainage_order)}
            drainage_colors['None'] = 'grey'
            return drainage_colors.get(row, 'grey')

        elif color_by == 'hydgrp':
            hydgrp_scale = {'A':0, 'A/B':0.5, 'B':1, 'B/C':1.5, 'C':2, 'C/D':2.5, 'D':3}
            val = hydgrp_scale.get(row, np.nan)
            if np.isnan(val):
                return 'grey'
            colors = px.colors.sequential.Blues
            idx = int(val / 3 * (len(colors)-1))
            return colors[idx]

        elif color_by == 'corcon':
            return {'None':'grey', 'Low':'green', 'Moderate':'yellow', 'High':'red'}.get(row, 'grey')

        else:  # compname
            unique_names = soils['compname'].dropna().unique()
            colors = px.colors.qualitative.Vivid
            comp_colors = {name: colors[i % len(colors)] for i, name in enumerate(unique_names)}
            return comp_colors.get(row, 'grey')

    polygons = []
    for _, soil_row in soils.iterrows():
        geom = soil_row.geometry
        polygon_color = get_color(soil_row[color_by])

        popup_content = html.Div([
            html.B("Component Name: "), soil_row['compname'], html.Br(),
            html.B("Drainage Class: "), soil_row['drainagecl'], html.Br(),
            html.B("Hydrologic Group: "), soil_row['hydgrp'], html.Br(),
            html.B("Corrosion Concrete: "), soil_row['corcon'], html.Br(),
        ])

        polys = geom.geoms if isinstance(geom, MultiPolygon) else [geom]
        for poly in polys:
            if poly.intersects(traffic_union):
                polygons.append(dl.Polygon(
                    positions=[(y, x) for x, y in poly.exterior.coords],
                    color=polygon_color, fill=True, opacity=0.6,
                    children=dl.Popup(popup_content)
                ))

    return polylines + polygons

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
import dash_leaflet as dl
import pandas as pd
from dash import html, dcc, Output, Input, State
from pyproj import Transformer
from shapely.geometry import MultiPolygon, Polygon
import plotly.express as px
import numpy as np

transformer = Transformer.from_crs("EPSG:3857", "EPSG:4326", always_xy=True)

county_options = MDTraffic['COUNTY_DES'].unique()

app = dash.Dash(__name__)

app.layout = html.Div([
    html.Div([
        html.Div([
            html.Label("County"),
            dcc.Dropdown(
                id='county-dropdown',
                options=[{'label': c, 'value': c} for c in county_options],
                value='MONTGOMERY', clearable=False
            ),
        ], style={'width': '18%', 'display': 'inline-block', 'padding-right': '10px'}),

        html.Div([
            html.Label("Street Name"),
            dcc.Dropdown(id='street-dropdown', multi=True),
        ], style={'width': '28%', 'display': 'inline-block', 'padding-right': '10px'}),

        html.Div([
            html.Label("Color Polygons By"),
            dcc.Dropdown(
                id='color-dropdown', clearable=False, value='drainagecl',
                options=[
                    {'label': 'Component Name', 'value': 'compname'},
                    {'label': 'Drainage Class', 'value': 'drainagecl'},
                    {'label': 'Hydrologic Group', 'value': 'hydgrp'},
                    {'label': 'Corrosion Concrete', 'value': 'corcon'}
                ]
            ),
        ], style={'width': '28%', 'display': 'inline-block', 'padding-right': '10px'}),

        html.Button('Execute', id='execute-button', n_clicks=0,
                    style={'width': '15%', 'display': 'inline-block', 'verticalAlign': 'bottom'}),
    ], style={'padding': '10px'}),

    dl.Map(id="map", style={'width': '100%', 'height': '600px'},
           zoom=10, center=(39.1547, -77.2405))
])

@app.callback(
    Output('street-dropdown', 'options'),
    Output('street-dropdown', 'value'),
    Input('county-dropdown', 'value')
)
def update_street_dropdown(county):
    streets = MDTraffic[MDTraffic['COUNTY_DES'] == county]['ROADNAME'].dropna().unique()[:100]
    return [{'label': s, 'value': s} for s in streets], []

@app.callback(
    Output("map", "children"),
    Input('execute-button', 'n_clicks'),
    State('county-dropdown', 'value'),
    State('street-dropdown', 'value'),
    State('color-dropdown', 'value')
)
def rebuild_map(n_clicks, county, selected_streets, color_by):
    if n_clicks == 0:
        return [dl.TileLayer()]

    traffic_df = MDTraffic[MDTraffic['COUNTY_DES'] == county]
    if selected_streets:
        traffic_df = traffic_df[traffic_df['ROADNAME'].isin(selected_streets)]
    traffic_4326 = traffic_df.to_crs(MDSoilPoly.crs)
    traffic_union = traffic_4326.unary_union

    polylines = [
        dl.Polyline(
            id=f"polyline-{n_clicks}-{idx}",
            positions=[transformer.transform(x, y)[::-1] for x, y, *_ in row.geometry.coords],
            color='blue',
            children=dl.Popup(html.Div([
                html.B("Road Name: "), row['ROADNAME'], html.Br(),
                html.B("AADT 2022: "), row['AADT_2022']
            ]))
        ) for idx, row in traffic_df.iterrows()
    ]

    soils = MDSoilPoly[MDSoilPoly.geometry.intersects(traffic_union)].copy()
    comps = MDSoilComp[MDSoilComp['majcompflag'] == 'Yes'][['mukey', 'compname', 'compkind', 'drainagecl', 'hydgrp', 'corcon', 'taxclname']]
    soils['mukey'] = soils['MUKEY'].astype(str)
    comps['mukey'] = comps['mukey'].astype(str)
    soils = soils.merge(comps, on='mukey', how='left')

    def get_color(row):
        if color_by == 'drainagecl':
            drainage_order = ['Excessively drained', 'Somewhat excessively drained', 'Well drained',
                              'Moderately well drained', 'Somewhat poorly drained', 'Poorly drained',
                              'Very poorly drained', 'Subaqueous']
            colors = px.colors.sequential.Plasma_r
            drainage_colors = {k: colors[int(i*(len(colors)-1)/(len(drainage_order)-1))]
                               for i, k in enumerate(drainage_order)}
            drainage_colors['None'] = 'grey'
            return drainage_colors.get(row, 'grey')

        elif color_by == 'hydgrp':
            hydgrp_scale = {'A':0, 'A/B':0.5, 'B':1, 'B/C':1.5, 'C':2, 'C/D':2.5, 'D':3}
            val = hydgrp_scale.get(row, 0)
            colors = px.colors.sequential.Blues
            idx = int(val/3*(len(colors)-1))
            return colors[idx]

        elif color_by == 'corcon':
            return {'None':'grey', 'Low':'green', 'Moderate':'yellow', 'High':'red'}.get(row, 'grey')

        else:  # compname
            unique_names = soils['compname'].dropna().unique()
            colors = px.colors.qualitative.Vivid
            comp_colors = {name: colors[i % len(colors)] for i, name in enumerate(unique_names)}
            return comp_colors.get(row, 'grey')

    polygons = []
    for idx, soil_row in soils.iterrows():
        geom = soil_row.geometry
        polygon_color = get_color(soil_row[color_by])

        popup_content = html.Div([
            html.B("Component Name: "), soil_row['compname'], html.Br(),
            html.B("Drainage Class: "), soil_row['drainagecl'], html.Br(),
            html.B("Hydrologic Group: "), soil_row['hydgrp'], html.Br(),
            html.B("Corrosion Concrete: "), soil_row['corcon'], html.Br(),
        ])

        polys = geom.geoms if isinstance(geom, MultiPolygon) else [geom]
        for poly_idx, poly in enumerate(polys):
            if poly.intersects(traffic_union):
                polygons.append(dl.Polygon(
                    id=f"polygon-{n_clicks}-{idx}-{poly_idx}",
                    positions=[(y, x) for x, y in poly.exterior.coords],
                    color=polygon_color, fill=True, opacity=0.6,
                    children=dl.Popup(popup_content)
                ))

    legend_items = sorted(set(soils[color_by].dropna()))
    legend = html.Div([
        html.Div([
            html.Span(style={'backgroundColor': get_color(val), 'width': '15px',
                             'height': '15px', 'display': 'inline-block', 'marginRight': '5px'}),
            html.Span(val)
        ]) for val in legend_items
    ], style={'position': 'absolute', 'top': '10px', 'right': '10px', 'backgroundColor': 'white',
              'padding': '10px', 'border': '1px solid #ccc', 'borderRadius': '5px', 'zIndex': '1000'})

    return [dl.TileLayer(), dl.LayerGroup(polylines + polygons), legend]

if __name__ == '__main__':
    app.run_server(debug=True)

```
